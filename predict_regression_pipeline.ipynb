# Predict based on given input

```python
import joblib
import pandas as pd

# Load model
loaded_model = joblib.load('best_model_regression_pipeline.pkl')

# Define input data (rows of features in order)
input_data = [
    [1,1.6,0.27,0,0.25,0.0108,2,0.2,1600,1600,42.8,7],
    [1,1.6,0.72,0,0.25,0.0283,2,0.2,1600,1300,42.8,7],
    [1,1.2,0.8,0,0.252,0.024,1.2,0.205,1600,1300,40,6],
    [0.7,1.2,0.8,0,0.32,0.016,4,0.308,1600,1300,42.8,6.5]
]

# Define column names
columns = [
    'cement_opc', 'scm_flyash', 'silica_sand', 'locally_avail_sand', 'w_b', 'hrwr_b',
    'perc_of_fibre', 'aspect_ratio', 'tensile_strength', 'density',
    'youngs_modulus', 'elongation'
]

# Create DataFrame
df_inputs = pd.DataFrame(input_data, columns=columns)

# Predict
predictions = loaded_model.predict(df_inputs)


# Print results
for i, pred in enumerate(predictions):
    print(f"Sample {i+1}: Predicted Compressive Strength = {pred:.2f}")
  
print("-------------------------------------------------")

# Print results with input row
for i, pred in enumerate(predictions):
    print(f"Sample {i+1}:")
    print(df_inputs.iloc[i])
    print(f"Predicted Compressive Strength = {pred:.2f}\n")
```
