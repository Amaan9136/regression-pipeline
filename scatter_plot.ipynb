```python
import pandas as pd
import matplotlib.pyplot as plt
import random

# Load dataset
df = pd.read_csv('data/new_refined.csv')

# Separate features and target
X = df.iloc[:, :-1]
y = df.iloc[:, -1]

# Create a color map for each feature
colors = ['#' + ''.join(random.choices('0123456789ABCDEF', k=6)) for _ in range(X.shape[1])]

# Create the plot
plt.figure(figsize=(10, 6))
for i, col in enumerate(X.columns):
    plt.scatter(X[col], y, label=col, alpha=0.7, color=colors[i])

plt.xlabel("Feature Values")
plt.ylabel(df.columns[-1])
plt.title("All Features vs Target")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
